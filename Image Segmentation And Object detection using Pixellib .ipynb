### **Image Segmentation & Detection Using Pixellib From Web Cam Video Stream**

```python
import cv2
import pixellib
from pixellib.instance import instance_segmentation
segment_image = instance_segmentation()
segment_image.load_model("mask_rcnn_coco.h5")  # loading model
cam = cv2.VideoCapture(0) # using webcam

while cam.isOpened(): # open my webcam
    res,frame=cam.read() # read frames from camera 
    result=segment_image.segmentFrame(frame,show_bboxes=True) # Apply Segmentation
    image=result[1] 
    cv2.imshow('Image Segmentation',image)

    if cv2.waitKey(10) & 0xFF==ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
